```python
# IQR function

def drop_outliers(aspmin_lite, field_name):
    iqr = 1.5 *(np.percentile(aspmin_lite[field_name],75) - np.percentile(aspmin_lite[field_name],25))
    aspmin_lite.drop(aspmin_lite[aspmin_lite[field_name]>(iqr+ np.percentile(aspmin_lite[field_name],75))].index,inplace=True)
    aspmin_lite.drop(aspmin_lite[aspmin_lite[field_name]<(np.percentile(aspmin_lite[field_name],25)-iqr)].index,inplace=True)
```

# <center> Hypothesis Testing </center>

##### Times of india dated Jan 18, 2019 states that "After doing your Computer Science Engineering  if you take up jobs as a Programming Analyst, software Engineer, Hardware engineer and  Associate Engineer you can earn upto 2.5 - 3 Lakhs as a fresher graduate."  Test this claim with the given data.

##### ReadMe

    * scipy.stats contains large number of probability distributions as well as a growing library of statistical functions.
    
    * scipy.stats.norm is normal continous random variable.
    
    * scipy.stats.t is a Students t continuous random variable.
    
    * math.sqrt is sqrt function.

```python
from scipy.stats import norm, t
from math import sqrt
```

```python
aspmin['DOJ'].value_counts()
```

Step - 1: 

* Alternate Hypothesis (Bold claim):$$ H_1:\ Earn \ more \ than \ 2.5 \ - \ 3.0 \ LPA $$
* Null Hypothesis (Status Quo):$$ H_0: \ Earn \ upto \ than \ 2.5 \ - \ 3.0 \ LPA $$

Step - 2:

* Collect a sample of size n = 20 $$ [350000, 350000, 120000, 215000, 2000000, 2050000, 245000, 120000, 210000, 305000, 320000, 315000, 330000, 85000, 150000, 300000, 240000, 350000, 180000, 350000] $$
* Compute sample mean$$ \bar{x} \ =  429250.0 $$


Step - 3:

* Compute Test Statistic:
$$ t = \frac{\bar{x} - \mu}{s/\sqrt[2]{n}} $$


Step - 4: 

* Decide $ \alpha $ or significance level 

With 95% confidence interval, $ \alpha $ = 1 - 0.85 = 0.15

Step - 5: 

* Step - 5.1: 
    *  2 tailed t-test : Failed to reject null hypothesis

* Step - 5.2: 
    * p-value = 0.6606

```python
# Two tail assuming 95% confidence interval

alpha = 1 - 0.85

t_critical_2 = t.ppf(1-alpha/2, df = 19)

print(t_critical_2)
```

```python
def t_score(sample_size,sample_mean, pop_mean, sample_std):
    numerator = sample_mean - pop_mean
    denomenator = sample_std / sample_size ** 0.5
    return numerator / denomenator
```

```python
# observations
samples = [350000, 350000, 120000, 215000, 2000000, 2050000, 245000, 120000, 210000, 305000, 320000, 315000, 330000, 85000, 150000, 300000, 240000, 350000, 180000, 350000]
```

```python
#sample mean and sample-size

sample_mean = sum(samples) / len(samples)
sample_size = len(samples)
```

```python
# calculate the sample_std 
score = 0
for i in samples:
    score += ((i - sample_mean) ** 2)/(sample_size - 1)
    sample_std = sqrt(score)
```

```python
print("Sample Standard Deviation",round(sample_std,3))
print("Sample Mean", sample_mean)
```

```python
# define population mean

pop_mean = (400000+350000)/2
print(pop_mean)

#Print t_score
t = t_score(sample_size, sample_mean, pop_mean, sample_std)

print("t_score is: ", t)
```

```python
print(min(samples), max(samples))
```

```python
# Ploting the sampling distribution with rejection regions
# Defining the x minimum and x maximum
x_min = 140000
x_max = 1660000

# Defining the sampling distribution mean and sampling distribution std
mean = pop_mean
std = sample_std / sample_size**0.5

plt.figure(figsize=(10,5))

# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)

# Computing the left and right critical values (Two tailed Test)
t_critical_left = pop_mean + (-t_critical_2 * std)
t_critical_right = pop_mean + (t_critical_2 * std)


# Shading the left rejection region
x1 = np.linspace(x_min, t_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='lightcoral')


# Shading the right rejection region
x2 = np.linspace(t_critical_right, x_max, 100)
y2 = norm.pdf(x2, mean, std)
plt.fill_between(x2, y2, color='goldenrod')

# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.000041))
```

```python
# Conclusion using t test

if(np.abs(t) > t_critical_2):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
# Conclusion using p test

p_value = 2 * (1.0 - norm.cdf(np.abs(t)))

print("p_value = ", round(p_value, 4))

if(p_value < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```
